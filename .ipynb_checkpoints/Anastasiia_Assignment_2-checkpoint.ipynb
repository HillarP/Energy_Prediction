```python
# Imorting librariess
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating a DataFrame object from a csv file
weather = pd.read_csv('weather.csv', encoding = 'latin_1')
weather #peaking at the first and last 5 rows
```

```python
#checking dimensions of the DataFrame
weather.shape
```

```python
#to further work with column headings, removing extra white spaces
# weather.rename(str.strip, axis='columns', inplace=True)
weather.columns = weather.columns.str.strip()

# checking missing valyes by columns
weather.isnull().sum()
```

```python
# We have found that values are completely missing in 'Evaporation (mm)', 'Sunshine (hours)', 
# '9am cloud amount (oktas)' and '3pm cloud amount (oktas)' so we can drop them 
weather.drop(columns=['Evaporation (mm)', 'Sunshine (hours)', '9am cloud amount (oktas)', '3pm cloud amount (oktas)'], inplace=True)
```

```python
#checking missing values by rows
weather.isnull().sum(axis=1)
```

```python
# row 174 contains 9 missing values, so it is relatively safe to drop it
weather.drop(index=174, inplace=True)
```

```python
# checking how many more missing values 
weather.isnull().values.sum()     #No more missing data
```

```python
# All values in 'Location' column seem identical, checking this assumption
weather['Location'].unique()
```

```python
# now we can drop the 'Location' column
weather.drop(columns=['Location'], inplace=True)
```

```python
#checking dtypes in each column
weather.dtypes
```

```python
# '9am wind speed (km/h)' values are object dtype as opposed to expected numeric;
# checking unique values of the column
weather['9am wind speed (km/h)'].unique()
```

```python
# Beaufort Wind Scale: 'Calm' equals to approx 0 km/h, replacing the value with numeric
weather['9am wind speed (km/h)'] = weather['9am wind speed (km/h)'].replace(['Calm'], '0')
# converting this column and '9am relative humidity (%)' column to float for consistency
weather = weather.astype({'9am wind speed (km/h)':'float', '9am relative humidity (%)':'float'})
weather.dtypes
```

```python
# Seting the DataFrame index using Date column
weather = weather.set_index('Date')
weather
```

```python
#checking statistical data
weather.describe()
```

```python
price_demand = pd.read_csv("price_and_demand.csv", header=None)
price_demand
```

```python
# no headers, so manually inserting them based on AEMO aggregated historical data (http://surl.li/jonpt)
#original headers are 'REGION','SETTLEMENTDATE','TOTALDEMAND','RRP','PERIODTYPE'
price_demand.columns =['Region','Date_Time','Total Demand','Price_per_MW','Period_Type']
price_demand
```

```python
# checking missing valyes by columns
price_demand.isnull().sum()    # no missing values
```

```python
#checking dtypes in each column - looks reasonable
price_demand.dtypes
```

```python
# All values in 'Region' and 'Period_Type' columns seem identical, checking this assumption
price_demand['Region'].unique()
price_demand['Period_Type'].unique()
# made a decision to drop these columns
price_demand.drop(columns=['Region', 'Period_Type'], inplace=True)
price_demand.head(5)
```

```python
# splittilg date&time into different columns and drop the original column
price_demand[['Date', 'Time']] = price_demand.Date_Time.str.split(expand = True)
price_demand.drop(columns=['Date_Time'], inplace=True)
#set the new 'Date' column as Index
price_demand = price_demand.set_index('Date')
price_demand
```

```python
# Checking the number of rows per each date
price_demand.index.value_counts()
```

```python
# there are 47 entries for 1/11/2022 compared to 48 on other dates 0:00 dataoint is missing, 
# so manually inserting them based on AEMO aggregated historical data (http://surl.li/jonpt)
missing_row = {"Total Demand": 4455.59, "Price_per_MW": 44.58, "Time": '0:00'}
missing_df = pd.DataFrame(missing_row, index=['1/11/2022'])
price_demand = pd.concat([missing_df, price_demand])


# also dropping row 24/04/2023 as it only has 1 value and not needed for a merge with weather df
price_demand.drop(index=['24/04/2023'], inplace=True)
price_demand.drop(columns=['Time'], inplace=True) #also removing Time column


# Finally checking the number of rows per each date
price_demand.index.value_counts()
```

```python
# Adding Max_Demand and Max_Price columns and assigning maximum value of respective columns for each date
price_demand['Max_Demand'] = price_demand.groupby(price_demand.index)['Total Demand'].transform('max')
price_demand['Max_Price'] = price_demand.groupby(price_demand.index)['Price_per_MW'].transform('max')

# Removing Total Demand and Price_per_MW as we are not going to analyze them
price_demand.drop(columns=['Total Demand', 'Price_per_MW'], inplace=True)
price_demand
```

```python
# removing dulicate rows
price_demand.reset_index().drop_duplicates(subset='index', keep='first').set_index('index')
```
